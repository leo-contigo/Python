## 막대 그래프
- plt.bar( x, color = 색상, width = 지정하지않으면 기본0.8, align='위치 ; 기본 center' )

```python
# 데이터 

member_id = ['m_01','m_02','m_03','m_04']
before_ex = [27, 35, 10, 55]
after_ex = [30, 50, 22, 73]

data = len(member_id)
index = np.arange(data)
plt.bar(index, before_ex)
plt.show();
```

```python
# x축 이름 member_id
plt.bar(index, before_ex, tick_label = member_id)
plt.show();
```

```python
# bar 색상 변경
color = ['y','g','b','m']
plt.bar(index, before_ex, tick_label = member_id, color = color)
plt.show();
```

```python
# bar 넓이 변경
color = ['y','g','b','m']
plt.bar(index, before_ex, tick_label = member_id, color = color, width = 0.5)
plt.show();
```

```python
# 가로 막대 plt.barh()
# width= 옵션 사용 안됨
# bar 색상 변경
color = ['y','g','b','m']
plt.barh(index, before_ex, tick_label = member_id, color = color)
plt.show();
```

```python
# 전, 후 데이터를 하나의 그래프로 

# bar두께 0.4
barWidth = 0.4

plt.bar(index, before_ex, color = 'g', align='edge', label = 'before', width = barWidth)
plt.bar(index+barWidth, after_ex, color = 'b', align='edge', label = 'after', width = barWidth)

plt.xticks(index + barWidth, member_id)
plt.xlabel('ID')
plt.ylabel('횟수')
plt.show()
```

### 히스토그램
- plt.hist(x, bin = 기본 10, 'auto')

```python
# 데이터가 어떤 분포를 갖는지 볼 때 이용.
# 히스토그램은 도수분포표를 막대그래프로 시각화.

# 도수분포표
# 변량 (variate) : 자료를 측정해 숫자로 표시한 것(예; 점수, 키 , 몸무게, 판매량, 시간 등..)
# 계급 (class) : 변량을 정해진 간격으로 나눈 구간 (예; 시험 점수를 50 ~ 60, 60 ~ 70, 구간으로 나눔)
# 계급의 간격 (class width) : 계급을 나눈 크기 (예; 위 간격 10)
# 도수 (frequency) : 나눠진 계급에 속하는 변량의 수
# 도수 분포표 (frequency distribution table) : 계급에 도수를 표시한 표
```

```python
# 데이터 

math = [76, 82, 84, 83, 90, 86, 85, 92, 72, 71, 100, 87, 81, 76, 94, 78, 81, 60, 79, 69, 74, 87, 82, 68, 79]
plt.hist(math)
```

```python
min(math), max(math)
```

```python
# 5간격으로 60 ~ 100 8개의 등급으로 나눠짐
plt.hist(math, bins = 8)
plt.show()
```

```python
plt.hist(math, bins = 8)
plt.xlabel('점수')
plt.ylabel('도수')
plt.grid()
plt.title('수학시험 히스토그램')
plt.show()
```

### 파이그래프

- plt.pie(x, labels = 항목 이름 , autopct= 비율 표시; '%.0f'- 정수/'%.1f'-실수, counterclock = False(시계방향))

```python
# 데이터
# 20명 학생에게 5개의 과일 중 제일 좋아하는 과일 선택

fruit = ['사과','바나나','딸기','오렌지','포도']
result = [7, 6, 3, 2, 2]

plt.pie(result);
```

```python
# 1:1 비율 설정
plt.figure(figsize = (5,5))
plt.pie(result);
```

```python
# 각 항목 표시
plt.figure(figsize = (5,5))
plt.pie(result, labels = fruit, autopct='%.1f');
```

```python
# 시계방향으로 표시
plt.figure(figsize = (5,5))
plt.pie(result, labels = fruit, autopct='%.1f', counterclock=False);
```

```python
# 특정 부분만 
plt.figure(figsize = (5,5))

explode = (0.1, 0, 0, 0, 0)
plt.pie(result, labels = fruit, autopct='%.1f', counterclock=False, explode=explode);
```
