```python
# 모듈 임포트 
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# 법정코드 요청 변수 확인 
# 개발 계정 상세보기 
# https://www.data.go.kr/iim/api/selectAPIAcountView.do
# [ 활용신청 상세기능정보 ] 에서 [확인] 버튼 클릭 후 요청변수 확인 
# 명세서 확인 - 지역코드 찾기 

# LAWD_CD 
# 각 지역별 코드 행정표준코드관리시스템(www.code.go.kr)의 법정동코드 10자리 중 앞 5자리
# [코드검색] - [법정명] 입력후 검색 
# 법정동명에 찾고자하는 동이름 입력후 검색 
# 법정동 결과 확인
# 2635010700 해운대 좌동

# 지역코드 5자리 하면... 해운대구 전체가 나오는디..?
```

```python
# Python3 샘플 코드 #

myKey = '개인 인증키 입력'
url = 'http://openapi.molit.go.kr/OpenAPI_ToolInstallPackage/service/rest/RTMSOBJSvc/getRTMSDataSvcAptTradeDev'

params ={'serviceKey' : myKey, 
         'pageNo' : '1', 
         'numOfRows' : '100',     # 결과 데이타 수 
         'LAWD_CD' : '26350',     # 법정명 지역코드 5자리
         'DEAL_YMD' : '202209' }  # 계약월

response = requests.get(url, params=params)
xml_str = response.text
xml_str
```

```python
# xml 텍스트 → 숩객체화 ('xm;'파싱 처리)
soup = BeautifulSoup(xml_str, 'xml')
# soup
```

```python
# 소스파악 및 타겟 리스트 설정
# item > 거래금액 +거래유형 + ....

target_list = soup.find_all('item')
len(target_list)
# target_list
```

```python
# 0번째 데이터 가져옴.
target_list[0]
```

```python
# 첫번째 요소 추출

# 아파트   apartment_name
# 건축년도 build_year
# 전용면적 area
# 층       floor
# 법정동   dong
# 거래금액 deal_amount
# 거래월   deal_day

apartment_name = target_list[0].아파트.text
build_year = target_list[0].건축년도.text
area = target_list[0].전용면적.text
floor = target_list[0].층.text
dong = target_list[0].법정동.text
deal_amount = target_list[0].거래금액.text.replace('    ', "")
deal_day = target_list[0].년.text + '.' + target_list[0].월.text + '.' + target_list[0].일.text


print(apartment_name, build_year, area, floor, dong, deal_amount, deal_day)
```

```python
# 데이터프레임

# 담을 빈 리스트
apt_list =[]

for item in target_list:
    apartment_name = item.아파트.text
    build_year = item.건축년도.text
    area = item.전용면적.text
    floor = item.층.text
    dong = item.법정동.text
    deal_amount = item.거래금액.text.replace('    ', "")
    deal_day = item.년.text + '.' + item.월.text + '.' + item.일.text
    
    apt_list.append([apartment_name, build_year, area, floor, dong, deal_amount, deal_day])

df = pd.DataFrame(apt_list, columns=['아파트명', '건축년도', '전용면적','층', '법정동','거래금액','거래월'])

df.sort_values('건축년도', ascending=False)
```
