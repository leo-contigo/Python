````python
# ì±„ë„ê´€ë¦¬ìê°€ ì˜¬ë ¤ë†“ì€ í”Œë ˆì´ë¦¬ìŠ¤íŠ¸ë¡œ ë°›ì•„ì˜´
playlists = youtube.playlists().list(
    channelId = channel_id,
    part = 'snippet',
    maxResults = 20).execute()
````

````python
import pandas as pd

ids = []
titles = []

for i in playlists['items']:
    ids.append(i['id'])
    titles.append(i['snippet']['title'])

pd.DataFrame([ids, titles]).T
````

````python
# ë°ì´í„° í”„ë ˆì„ìœ¼ë¡œ ã„±ã„±   
a = pd.DataFrame([ids,titles]).T
a.columns=['PlayLists','Titles']
a.head()

# ì¬ìƒëª©ë¡ id ê°’ í™•ì¸ ê°€ëŠ¥!
````

````python
# ì˜ìƒ ì„ íƒí•˜ê¸°
# ì¶œì¥ ì‹­ì˜¤ì•¼ã…‹ã…‹ã…‹ã…‹ã…‹ ì„ íƒ
# PlayLists 4ë²ˆ 

ya = a['PlayLists'][4]
video = youtube.playlistItems().list(
    playlistId = ya,
    part = 'snippet',
    maxResults = 50,
)
````

````python
video_list = video.execute()

video_names=[]
video_ids=[]
date=[]

for v in video_list['items']:
    video_names.append(v['snippet']['title'])
    video_ids.append(v['snippet']['resourceId']['videoId'])
    date.append(v['snippet']['publishedAt'])
    
pd.DataFrame([date,video_names,video_ids]).T
````

````python
a2 = pd.DataFrame([date,video_names,video_ids]).T
a2.columns=['Date','Title','IDS']
a2.head()
````
