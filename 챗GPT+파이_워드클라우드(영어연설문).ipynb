```python
# pip list
# wordcloud와 matplotlib설치 확인함.
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# 주어진 연설문 텍스트
speech = """
영어 연설문 겁나 김..
결과 내고 생략
"""

# 워드 클라우드 생성
wordcloud = WordCloud(width=800, height=800, background_color='white').generate(speech)

# 워드 클라우드 그리기
plt.figure(figsize=(8, 8), facecolor=None)
plt.rc('font' , family='NanumBarunGothic')
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad=0)

plt.show()
```

## 둥근원 모양으로

```python
from wordcloud import WordCloud, ImageColorGenerator
import numpy as np
from PIL import Image
```

```python
# 이미지 마스크 로드
mask1 = np.array(Image.open("/content/dry-clean.png"))  # 원하는 모양 이미지 파일 경로

# 워드 클라우드 생성
wordcloud = WordCloud(width=500, height=500, background_color='white', mask=mask1).generate(speech)

# 워드 클라우드 그리기
plt.figure(figsize=(8, 8), facecolor=None)
plt.rc('font' , family='NanumBarunGothic')
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad=0)

plt.show()
```
